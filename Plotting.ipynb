```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'Age': [23,25,22,30,35,40,56],
    'Height' : [165,170,160,175, 180, 120, 156],
    'Weight' : [65, 70, 60, 75, 80, 58, 68],
    'Gender' : ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male'],
    'Smoker' : ['No', 'Yes', 'No', 'No', 'No', 'Yes', 'Yes'],
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
# UnderStand the Data (Step  1)
print("Data Summary\n",df.describe())
print()
print("Data Types",df.dtypes)
```

```python
#Step 2 Unvariate Analysis - Numerical Data
#Age (Discrete Data)
plt.figure(figsize = (10, 4))
plt.subplot(1,2,1)
sns.histplot(df['Age'], kde = True, color = 'blue') # kde is kernal density
plt.title('Age Distribution (Discrete Data)')

#Height (Continous data)
plt.figure(figsize= (10,4))
plt.subplot(1,2,1)
sns.histplot(df['Height'], kde = True, color = 'green')
plt.title('Height distribution')
plt.tight_layout()
```

```python
#Step 3
plt.figure(figsize = (10, 4))

plt.subplot(1,2,1)
sns.countplot(x='Gender', data =df, palette = 'Set1')
plt.title('Gender Count (Categorical Data)')

plt.subplot(1,2,2)
sns.countplot(x='Smoker', data =df, palette = 'Set2')
plt.title('Somker Count (Categorical Data)')

plt.tight_layout()
plt.show()
```

```python
#Step 4 Bivariate analysis - numeric data
sns.scatterplot(x = 'Age', y='Weight', hue = 'Gender', data = df)
plt.title('Scatter plot (Age vs Weight)')
plt.show()
```

```python
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Smoker'], y=df['Height'], palette='Set3')
plt.title("Outlier in Height by Smoker Status")
```

```python
corr_matrix = df[['Age', 'Height', 'Weight']].corr()

# Step 2: Create the heatmap
plt.figure(figsize=(10, 4))

# Heatmap of correlation matrix
plt.subplot(1, 2, 1)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```

```python
df['Height_m'] = df['Height'] / 100

# Calculate BMI
df['BMI Index'] = df['Weight'] / (df['Height_m'] ** 2)
```
